```python
import pdfplumber
import pandas as pd

data = []

with pdfplumber.open("dbs_statement.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        for line in text.split('\n'):
            # Adjust this regex to match your bank's line format
            if line.startswith('202'):
                parts = line.split()  # crude split; better to use regex
                date = parts[0]
                description = " ".join(parts[1:-1])
                amount = parts[-1]
                data.append([date, description, amount])

df = pd.DataFrame(data, columns=["Date", "Description", "Amount"])
print(df.head())
```

```python
import pdfplumber
import pandas as pd

all_data = []

with pdfplumber.open("oub_statement.pdf") as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            for row in table:
                if row and any(cell is not None for cell in row):
                    all_data.append(row)

df = pd.DataFrame(all_data)
print(df.head(10))
```

```python
import pdfplumber
import pandas as pd

from pdf2image import convert_from_path
import pytesseract
import pandas as pd

pages = convert_from_path("dbs_statement.pdf", dpi=300)

for i, page in enumerate(pages):
    text = pytesseract.image_to_string(page)
    print(f"--- Page {i+1} ---\n{text}\n")
```
